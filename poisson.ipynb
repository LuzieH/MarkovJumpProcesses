```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial
```

### Poisson process

```python
rate = 0.5
N_jumps = 10
S = np.random.exponential(1/rate, N_jumps)
S
```

```python
J = np.cumsum(S)
J
```

```python
plt.step(J, range(0,N_jumps))
plt.xlabel('t')
plt.ylabel('$X_t$')
plt.title("Poisson process")
```

### Independent calls to a call-center

```python
N = 10 # nbr people
p = 2/N # call probability on a given day
constant = p*N # lambda constant
```

```python
N_samples = 5
calls = np.random.binomial(N, p, N_samples)
calls
```

```python
plt.figure(1)
xs = np.arange(0,np.minimum(10,N),1)
bins = xs-0.5
#m = plt.hist(calls,density=True,bins=bins,label="Binomial distribution")
poiss = [np.exp(-constant)*constant**x/factorial(x) for x in xs][:-1]
binomial = [factorial(N)/(factorial(x)*factorial(N-x))*p**x*(1-p)**(N-x) for x in xs][:-1]
plt.stairs(poiss, bins,label="Poisson distribution")
plt.stairs(binomial, bins,label="Binomial distribution")
plt.legend()
```
