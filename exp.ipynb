```python
import numpy as np
import matplotlib.pyplot as plt
```

### Exponential random variables

```python
rate = 2.
N=10
draws = np.random.exponential(1/rate, N)
draws
```

```python
mean = np.mean(draws)
mean
```

```python
plt.figure(1)
plt.hist(draws,density=True)
xs = np.linspace(0,np.max(draws),100)
plt.plot(xs,rate*np.exp(-rate*xs),label="exponential probability density function")
plt.title("Histogram")
plt.legend()
```
